```python
def plot_top_10_frequent(df):
    for cols in df.columns:
        plt.figure(figsize=(6,6))
        df[cols].value_counts().head(10).plot(kind='bar')
        plt.xlabel(cols)
        plt.ylabel("Counts")
        plt.title(f"Top Frequent {cols}")
        plt.show()
```

```python
df.dropna(inplace=True)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
oversampler = RandomOverSampler(sampling_strategy='auto', random_state=42)
```

```python
X=df.drop('Dangerous',axis=1)
y=df['Dangerous']
```

```python
new_X,new_y=oversampler.fit_resample(X,y)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',sparse=False)
```

```python
X_encoded=ohe.fit_transform(new_X)
```

```python
columns=ohe.get_feature_names_out(['AnimalName', 'symptoms1', 'symptoms2', 'symptoms3', 'symptoms4',
       'symptoms5'])
```

```python
X_encoded=pd.DataFrame(X_encoded,columns=columns)
```

```python
X_encoded
```

```python
encoded_y=[1 if i=='Yes' else 0 for i in new_y]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_encoded,encoded_y,test_size=0.2,random_state=2)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```
